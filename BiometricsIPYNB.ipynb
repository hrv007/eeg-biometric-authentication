```python
import pandas as pd
import numpy as np
import scipy
from scipy.io import loadmat
import matplotlib.pyplot as plt
import tensorflow as tf
from librosa.feature import mfcc, melspectrogram
from librosa.display import specshow
import librosa
```

```python
df = pd.read_csv("eeg-biometric-competition/Enrollment_Info.csv")
columns = df.columns
df.head()
```

```python
subs = {}
for subid in df['subject'].unique():
    if subid == 'sub021':
        break
    subs[subid] = []
    #os.mkdir('Mel Spectrograms/'+subid)
```

```python
for epochid, subid in zip(df['EpochID'], df['subject']):
    if subid == 'sub021':
        break
    subs[subid].append(epochid)
```

```python
epochid = 'epoch000101'
annots = loadmat('eeg-biometric-competition/Enrollment/'+epochid+'.mat')
wave = annots['epoch_data']
wave

wave = wave.flatten()

mel_spec = librosa.feature.melspectrogram(y=wave, n_mels=128)

mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

plt.figure(figsize=(14, 8))
specshow(mel_spec_db, y_axis='log', fmax=8000, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
plt.savefig('Mel Spectrograms/sub001/'+epochid+'.png')
plt.close()
```

```python
for subid in subs.keys():
    for epochid in subs[subid]:
        annots = loadmat('Enrollment/'+epochid+'.mat')
        wave = annots['epoch_data']

        wave = wave.flatten()

        mel_spec = librosa.feature.melspectrogram(y=wave, n_mels=128)

        mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

        plt.figure(figsize=(14, 8))
        specshow(mel_spec_db, y_axis='log', fmax=8000, x_axis='time')
        plt.colorbar(format='%+2.0f dB')
        plt.title('Mel spectrogram')
        plt.tight_layout()
        plt.savefig('Mel Spectrograms/'+subid+'/'+epochid+'.png')
        plt.close()
```
